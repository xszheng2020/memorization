```python
matplotlib.rcParams.update({'font.size': 24})

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))

x = range(10, 100, 10)

results = list((np.array(results_1)+np.array(results_2)+np.array(results_3))/3)

ax.errorbar(x, results, 
            yerr=None, 
            linestyle='-', 
            fmt='o', label='random attribution', color='tab:green', capsize=5, 
#             alpha=0.6
           )

ax.errorbar(x, results_0, 
            yerr=None, 
            linestyle='-', 
            fmt='o', label='mem attribution', color='tab:orange', capsize=5, 
#             alpha=0.6
           )

ax.legend(fontsize=20)



ax.grid(True)

ax.set_xlabel("Fraction Removed (%)")
ax.set_xticks(x)

ax.set_ylabel("Reduction Rate (%)")
# ax.set_yticks(np.arange(0, 100.00+10, 10))
ax.set_yticks(np.arange(np.round(min(results+results_0), -1), np.round(max(results+results_0), -1)+10, 10))

filename = "saved/vis/Eval_Attr_SST.pdf"
os.makedirs(os.path.dirname(filename), exist_ok=True)  

plt.savefig(filename, bbox_inches='tight', pad_inches=0.1)  
```
