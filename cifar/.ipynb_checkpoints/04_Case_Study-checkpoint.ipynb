```python
for i in range(10):
    if i==0:
        tmp = df_0[df_0['label']==i].sort_values(by=['influence'], ascending=False).head(1)
    else:
        tmp = tmp.append(df_0[df_0['label']==i].sort_values(by=['influence'], ascending=False).head(1))
tmp[['cifar_index', 'label', 'prediction']].head(10)
```

```python
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]
def unnormalize(img, mean, std):
    for t, m, s in zip(img, mean, std):
        t.mul_(s).add_(m) # unnormalize
    npimg = img.numpy()
    npimg = np.transpose(npimg, (1, 2, 0))
    
    return npimg
```

```python
import matplotlib.colors as colors
```

```python
matplotlib.rcParams.update({'font.size': 20})

fig, axs = plt.subplots(2, 10, 
                        figsize=(16, 4)
                       )


for idx in range(len(tmp)):
    outputs = tmp.iloc[idx]
    
    original_image = unnormalize(train_dataset_full[outputs['cifar_index']][0], mean, std)
    attributions = zoom(outputs['attributions'], (32, 32), order=2)

    axs[0, idx].imshow(original_image)
    axs[0, idx].set_title(classes[outputs['label']])
    axs[0, idx].set_axis_off()
    
    norm = colors.DivergingNorm(vmin=min( attributions.min(), 
                                     -(attributions.min()) ), 
                            vcenter=0, 
                            vmax=max(attributions.max(), 
                                     -(attributions.max()) )
                           )
    Y = 0.299 * original_image[:, :, 0] + 0.587 * original_image[:, :, 1] + 0.114 * original_image[:, :, 2]
    axs[1, idx].imshow(Y, cmap='gray', vmin=0, vmax=1, alpha=1)
    attr_img = axs[1, idx].imshow(attributions, 
#                          cmap=cmap, 
#                          vmin=-0.8, vmax=0.8,
                         cmap='RdBu_r', 
                         norm=norm, 
                         alpha=0.8)    
    axs[1, idx].set_axis_off()
    
fig.tight_layout(h_pad=-1, w_pad=1)
# plt.tight_layout()
# plt.show()
plt.savefig('saved/vis/Vis_CIFAR_10_Attr.pdf', bbox_inches='tight', pad_inches=0.1)  
```
