```python
def get_random_attribution(attributions):
    
    length = len(attributions)
        
    return np.random.rand(length)     
```

```python
def get_mask(percentage, attributions):
    
    length = len(attributions)
    mask = np.zeros(length)
    
    attributions = attributions.copy()
        
    k = int( percentage/100*(length-2) )
    
    if k==0:
        k = 1

    attributions[0] = -1e30 # [CLS]
    attributions[-1] = -1e30 # [SEP]
        
    indices = (-attributions).argsort()[:k]
    
    for i in indices:
        mask[i] = 1
            
    mask[0] = 0 # [CLS]
    mask[-1] = 0 # [SEP]
        
    return mask
```
